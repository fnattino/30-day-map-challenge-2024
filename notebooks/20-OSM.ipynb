# #20 - Data: OpenStreetMap

This map is based on the output of the [City River Spaces (CRiSp) project](https://research-software-directory.org/projects/crisp).

```python
import pathlib

import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
data_dir = pathlib.Path("../data")
fig_dir = pathlib.Path("../figs")
```

```python
# city_boundary = gpd.read_file(data_dir / "city_boundary_Bucharest.gpkg")
highways = gpd.read_file(data_dir / "highways_Bucharest.gpkg")
railways = gpd.read_file(data_dir / "railways_Bucharest.gpkg")
waterways = gpd.read_file(data_dir / "waterway_Dambovita.gpkg")
corridor = gpd.read_file(data_dir / "corridor_Dambovita.gpkg")
segments = gpd.read_file(data_dir / "segments_Dambovita.gpkg")
```

```python
xmin, ymin, xmax, ymax = corridor.geometry[0].buffer(100).bounds
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 10))

waterways.geometry[[1]].plot(ax=ax, color="blue")
highways.plot(ax=ax, color="grey", lw=0.3)
railways.plot(ax=ax, color="grey", lw=0.3)
# city_boundary.boundary.plot(ax=ax, color="yellow", alpha=0.5, lw=2)
segments.boundary.plot(ax=ax, color="red", lw=3)
corridor.geometry[[0]].plot(ax=ax, color="red", alpha=0.2, lw=5)
ax.set_xlim(xmin, xmax)
ax.set_ylim(ymin, ymax)
ax.set_axis_off()

plt.savefig(fig_dir / "20-OSM-CRiSp.png", dpi=300)
```
